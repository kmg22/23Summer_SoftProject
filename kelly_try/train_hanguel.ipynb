```python
import numpy as np
import os

os.environ['CUDA_VISIBLE_DEVICES'] = '1'
os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'

actions = ['ㄱ','ㄴ','ㄷ','ㄹ','ㅁ','ㅂ','ㅅ','ㅇ','ㅈ','ㅊ','ㅋ','ㅌ','ㅍ','ㅎ','ㅏ','ㅑ','ㅓ','ㅕ','ㅗ','ㅛ','ㅜ','ㅠ','ㅡ','ㅣ','ㅐ','ㅒ','ㅔ','ㅖ','ㅚ','ㅟ','ㅢ','space','backspace','1','2','3']


data = np.concatenate([
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㄱ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㄴ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㄷ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㄹ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅁ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅂ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅅ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅇ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅈ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅊ_1689329536.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅋ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅌ_1689329907.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅍ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅎ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅏ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅑ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅓ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅕ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅗ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅛ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅜ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅠ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅡ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅣ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅐ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅒ_1689329907.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅔ_1689329907.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅖ_1689329907.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅚ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_ㅟ_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_space_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_backspace_1689328397.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_1_1689329907.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_2_1689330070.npy'),
    np.load('C:/Users/kelly/IdeaProjects/Project mid/dataset/seq_3_1689330070.npy')
   
], axis=0)

data.shape

################################################

x_data = data[:, :, :-1]
labels = data[:, 0, -1]

print(x_data.shape)
print(labels.shape)

################################################

from tensorflow.keras.utils import to_categorical

y_data = to_categorical(labels, num_classes=len(actions))
y_data.shape

################################################

from sklearn.model_selection import train_test_split

x_data = x_data.astype(np.float32)
y_data = y_data.astype(np.float32)

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.1, random_state=2021)

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)

################################################

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential([
    LSTM(32, activation='relu', input_shape=x_train.shape[1:3]),
    Dense(1024, activation='relu'),
    Dense(len(actions), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
model.summary()

################################################

from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau

history = model.fit(
    x_train,
    y_train,
    validation_data=(x_val, y_val),
    epochs=200,
    callbacks=[
        ModelCheckpoint('models/models.h5', monitor='val_acc', verbose=1, save_best_only=True, mode='auto'),
        ReduceLROnPlateau(monitor='val_acc', factor=0.5, patience=50, verbose=1, mode='auto')
    ]
)
```
